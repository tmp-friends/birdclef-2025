##  Move submission csv

```python
import glob
import shutil
import os

# 検索するパスパターンを指定
pattern = '/kaggle/working/output/infer/**/*/submission.csv'

# `glob`を使って指定パターンにマッチするファイルを探す
files = glob.glob(pattern)

# ファイルが見つかったかどうかをチェック
if files:
    # 見つかったファイルを取得
    submission_file = files[0]
    
    # 移動先のディレクトリとファイル名を設定
    destination = '/kaggle/working/submission.csv'
    
    # ファイルを移動
    shutil.move(submission_file, destination)
    print(f"Moved {submission_file} to {destination}")
else:
    print("No submission.csv file found.")
```
